```python
#loading libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/claricekim/Downloads/music_genre.csv")
df
```

```python
df2=df[(df.music_genre == "Jazz") | (df.music_genre == "Classical")]
df=df2
df
```

```python
#I imported the "LabelEncoder" function from sklearn library. 
import sklearn.preprocessing
from sklearn.preprocessing import LabelEncoder
#Then, I renamed it to "LE" for ease of typing. 
LE = LabelEncoder()

#"LabelEncoder" allows for my target value to be converted from letters to numbers, which the code can interpret. 
#Specifically, my target values are under the 'label' column, and they are the musical genres. 
#In the code below, I rename the columns with new target values ('Electronic'=0, 'Anime'=1, etc.) as "target"

target = LE.fit_transform(df['music_genre'])
#This drops the 'label' column from the dataset
df.drop('music_genre', axis=1, inplace=True)
#This adds the column with new target values back into the dataframe
df['music_genre'] = target
#This displays the modified dataframe with new target values.
df
```

```python
#The dataset needs to be modified to include only numerical values. 
#I will assign numerical values to the "key" and "mode" features. 
#I will drop the rest of the columns without numerical values as well as the numerical id of the song that will not help classify it into a genre
X=df.drop(['instance_id','artist_name','track_name', 'obtained_date'],axis=1)
```

```python
#Modifying the "key" feature

key = LE.fit_transform(X['key'])
#This drops the 'label' column from the dataset
X.drop('key', axis=1, inplace=True)
#This adds the column with new target values back into the dataframe
X['key'] = key
#This displays the modified dataframe with new target values.
X
```

```python
#Modifying the "mode" feature

mode = LE.fit_transform(X['mode'])
#This drops the 'label' column from the dataset
X.drop('mode', axis=1, inplace=True)
#This adds the column with new target values back into the dataframe
X['mode'] = mode
#This displays the modified dataframe with new target values.
X
```
